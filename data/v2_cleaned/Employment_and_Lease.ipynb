```python
df2 = pd.read_csv('/data_gov_lse_rexus.csv', on_bad_lines='skip')
df2
```

```python
df2['Lessor State'] = df2['Lessor State'].replace('Illinois', 'IL')

states_raw = df2
filter1 = states_raw['Lessor Country'] == 'US'

states_no_ca = states_raw[filter1]
states_no_ca.describe(include='all')
filter2 = states_no_ca['Lessor State'].isna()

states_us = states_no_ca[~filter2]
states_us
```

```python
df = pd.read_csv('data_gov_bldg_rexus.csv')
zip_to_county = df[['Bldg County', 'Bldg Zip']]

df2['Lessor Zip Code'] = df2['Lessor Zip Code'].astype(str)
zip_to_county['Bldg Zip'] = zip_to_county['Bldg Zip'].astype(str)

merged = df2.merge(zip_to_county, left_on='Lessor Zip Code', right_on='Bldg Zip')

just_what_i_want = merged.loc[:, ['Bldg County', 'Lease Annual Rent Amount', 'Lease ANSI Rentable Sqft']]
just_what_i_want.describe()

states = states_us.loc[:, ['Lessor State', 'Lease Annual Rent Amount', 'Lease ANSI Rentable Sqft']]

avg = just_what_i_want.groupby('Bldg County').mean()
state_avg = states.groupby('Lessor State').mean()

avg['Average Rent per Sqft'] = avg['Lease Annual Rent Amount'] / avg['Lease ANSI Rentable Sqft'].replace(0, np.nan)
state_avg['Average Rent per Sqft'] = state_avg['Lease Annual Rent Amount'] / state_avg['Lease ANSI Rentable Sqft'].replace(0, np.nan)
avg.to_csv('Lease_cost_by_county.csv')
state_avg.to_csv('Lease_cost_by_state.csv')
```
