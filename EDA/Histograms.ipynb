```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
leases = pd.read_csv('Lease_cost_by_county.csv')
leases_state = pd.read_csv('Lease_cost_by_state.csv')

sns.set_style('white')
ax = sns.histplot(data= leases['Average Rent per Sqft'], binwidth = 10, color = '#228B22', edgecolor = '#228B22')
ax.set_xlabel('Average Rent per Sqft')
ax.set_ylabel('Count')
ax.set_title('Distribution of Average Rent per Sqft by County')
ax.grid(True, axis="y", linestyle="--", color="gray", alpha=0.1)

fig = ax.get_figure()
fig.savefig("County.png", dpi=300, bbox_inches='tight')
```

```python
sns.set_style('white')
ax = sns.histplot(data= leases_state['Average Rent per Sqft'], binwidth = 3, color = '#D6853E', edgecolor = '#D6853E')
ax.set_xlabel('Average Rent per Sqft')
ax.set_ylabel('Count')
ax.set_title('Distribution of Average Rent per Sqft by State')
ax.grid(True, axis="y", linestyle="--", color="gray", alpha=0.1)

fig = ax.get_figure()
fig.savefig("State.png", dpi=300, bbox_inches='tight')
```
