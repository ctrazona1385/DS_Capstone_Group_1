```python
import pandas as pd
```

```python
df_cdc = pd.read_csv(r"C:\Users\17063\OneDrive\Documents\GitHub\DS_Capstone_Group_1\data\v1_renamed\census_county_data.csv")
df_hpsa = pd.read_csv(r"C:\Users\17063\OneDrive\Documents\GitHub\DS_Capstone_Group_1\data\v1_renamed\hpsa_county_composite_score.csv")
df_sahie = pd.read_csv(r"C:\Users\17063\OneDrive\Documents\GitHub\DS_Capstone_Group_1\data\v2_cleaned\sahie_cleaned.csv")
```

```python
print(df_hpsa.shape)
print(df_cdc.shape)
```

```python
# cast both to string
df_cdc['geoid'] = df_cdc['geoid'].astype(str)
df_hpsa['geoid'] = df_hpsa['geoid'].astype(str)
df_sahie['geoid'] = df_sahie['geoid'].astype(str)
df_cdc["geoid"] = df_cdc["geoid"].astype(str).str.zfill(5)
df_hpsa["geoid"] = df_hpsa["geoid"].astype(str).str.zfill(5)
df_sahie["geoid"] = df_sahie["geoid"].astype(str).str.zfill(5)


```

```python
merged_df = pd.merge(df_cdc, df_hpsa, on='geoid', how='left')
```

```python
all_merged_df = pd.merge(merged_df, df_sahie, on='geoid', how='left')
```

```python
all_merged_df.fillna(0, inplace=True)
```
