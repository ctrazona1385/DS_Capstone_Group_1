```python
# --- Cleaning steps for all_merged.csv ---

# 1. Drop Connecticut rows (geoid starts with '09')
df = df_all_merged[~df_all_merged['geoid'].astype(str).str.startswith('09')].copy()

# 2. Drop hpsa_designation_count column
df = df.drop(columns=['hpsa_designation_count'])

# 3. Fill nulls with 0 for HPSA / physician columns
fill_zero_cols = ['hpsa_designated', 'md_primary_care_count', 'do_primary_care_count', 'pcp_count']
df[fill_zero_cols] = df[fill_zero_cols].fillna(0)

# 4. Impute missing total_population using poverty columns
#    total_population = total_population_poverty / (poverty_rate_pct / 100)
missing_pop = df['total_population'].isnull()
df.loc[missing_pop, 'total_population'] = (
    df.loc[missing_pop, 'total_population_poverty'] /
    (df.loc[missing_pop, 'poverty_rate_pct'] / 100)
)

# 5. Recalculate pcp_per_100k with imputed population
df['pcp_per_100k'] = df['pcp_count'] / df['total_population'] * 100_000

# 6. Save cleaned CSV
df.to_csv(r"C:\Users\17063\OneDrive\Documents\GitHub\DS_Capstone_Group_1\data\v2_cleaned\all_merged.csv", index=False)
```

```python
# --- Data Quality Report ---
print("=== Data Quality Report ===")
print(f"Rows: {len(df):,}  |  Columns: {df.shape[1]}")

print(f"\nConnecticut rows remaining: {df['geoid'].astype(str).str.startswith('09').sum()}")
print(f"'hpsa_designation_count' column present: {'hpsa_designation_count' in df.columns}")

print("\nNull counts (columns with any nulls):")
null_counts = df.isnull().sum()
null_counts = null_counts[null_counts > 0]
print(null_counts.to_string() if not null_counts.empty else "  None — no nulls remaining")

print("\npcp_per_100k descriptive stats:")
print(df['pcp_per_100k'].describe().round(2))
print(f"Inf values in pcp_per_100k: {(df['pcp_per_100k'] == float('inf')).sum()}")

print("\nImputation summary:")
pre_ct_filter = df_all_merged[~df_all_merged['geoid'].astype(str).str.startswith('09')]
for col in ['hpsa_designated', 'md_primary_care_count', 'do_primary_care_count', 'pcp_count']:
    n = pre_ct_filter[col].isnull().sum()
    print(f"  {col}: {n:,} nulls filled with 0")
print(f"  total_population: {pre_ct_filter['total_population'].isnull().sum():,} rows imputed from poverty data")
```
